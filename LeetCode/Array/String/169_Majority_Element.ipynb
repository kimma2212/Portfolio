```python
#Original approach
#O(n) / O(n)


from typing import List




def majorityElement(nums: List[int]) -> int:
    counter = {}
    for num in nums:
        if num in counter:
            counter[num] += 1
        else:
            counter[num] = 1
    


    return max(counter, key=counter.get)




nums = [3,2,3]

print(majorityElement(nums))
```

```python
from typing import List

#If the element is majority, when it is sorted, it will always be at the middle index
#O(N log N) / O(1)


def majorityElement(nums: List[int]) -> int:
    nums.sort()
    return nums[len(nums) // 2]




nums = [3,2,3]

print(majorityElement(nums))
```

```python
from typing import List


#Boyer-Moore Voting Algorithm
#O(N) / O(1)

def majorityElement(nums: List[int]) -> int:
    candidate = None
    count = 0

    for i in nums:
        if count == 0:
            candidate = i
        
        if i == candidate:
            count += 1
        else:
            count -= 1

    return candidate
        



nums = [3,2,3]

print(majorityElement(nums))
```
