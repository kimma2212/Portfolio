```python
#설계도, 기차칸

class ListNode:
    def __init__(self, val):
        self.val = val
        self.next = None
```

```python
node1 = ListNode(1)
node2 = ListNode(2)
node3 = ListNode(3)
```

```python
#Connect
node1.next = node2
node2.next = node3
```

```python
current_node = node1

while current_node is not None:
    print(current_node.val, end=" -> ")
    current_node = current_node.next

print("END")
```
