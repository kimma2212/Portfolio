```python
import pandas as pd

# Load the weather data
weather_data = pd.read_csv("data/weather/7_12_22.csv")

# Filter the data for the required date range
filtered_weather_data = weather_data.iloc[779:962]

# Create a new 'Month' column based on conditions
filtered_weather_data['Month'] = filtered_weather_data['DATE'].apply(
    lambda x: 1 if '-01-' in x else
              2 if '-02-' in x else
              3 if '-03-' in x else
              4 if '-04-' in x else
              5 if '-05-' in x else
              6 if '-06-' in x else 
              7 if '-07-' in x else 
              8 if '-08-' in x else 
              9 if '-09-' in x else 
              10 if '-10-' in x else 
              11 if '-11-' in x else 
              12 if '-12-' in x else None
)

# Select the required columns
selected_data = filtered_weather_data[['Month', 'PRCP', 'TMAX', 'TMIN', 'TOBS']]

# Create a new DataFrame
new_pd = pd.DataFrame(selected_data)



# Print the new DataFrame to verify
print(new_pd.head())


new_pd.to_csv("data/weather/modified_weather_7_12_2022.csv", index=False)
```

```python
import pandas as pd

# Load original data
df1 = pd.read_excel('combined_corn.xlsx')

# Load weather data
df2 = pd.read_csv("data/weather/modified_weather_1_6_2022.csv")
df3 = pd.read_csv("data/weather/modified_weather_7_12_2022.csv")


df4 = pd.read_csv("data/merged_bean_22.csv")




def fill_with_avg(series):
    return series.fillna((series.ffill() + series.bfill()) / 2)

df3['Month'] = fill_with_avg(df3['Month'])
df2['Month'] = fill_with_avg(df2['Month'])
df1['Month'] = fill_with_avg(df1['Month'])
df4['Month'] = fill_with_avg(df4['Month'])



# Ensuring 'Month' column is consistent and of the same type in all dataframes
df1['Month'] = df1['Month'].astype(int)
df2['Month'] = df2['Month'].astype(int)
df3['Month'] = df3['Month'].astype(int)
df4['Month'] = df4['Month'].astype(int)

# Concatenate weather data
weather_data = pd.concat([df2, df3])



# Merge original data with weather data using an outer merge
result = pd.merge(df1, weather_data, how='outer', on='Month')

# # Inspect the result
# print(result.head())
# print(result.tail())


result = pd.merge(df4, result, how='outer', on='Month')


# Save the merged result
result.to_csv("data/merged_2022.csv", index=False)
```

```python
import pandas as pd

# 데이터 로드
dfa = pd.read_csv("selected_data.csv")
dfb = pd.read_csv("data/merged_bean_22.csv")



# 월별 상위 30개 데이터 추출 및 일련번호 추가
dfa_30 = dfa.groupby('Month').head(30).reset_index()
dfa_30['Seq'] = dfa_30.groupby('Month').cumcount() + 1

dfb_30 = dfb.groupby('Month').head(30).reset_index()
dfb_30['Seq'] = dfb_30.groupby('Month').cumcount() + 1

# dfb의 'Price' 열 이름 변경
dfb_30.rename(columns={'Price': 'Bean_Price'}, inplace=True)

# 데이터 병합: 'Month'와 'Seq'를 기준으로 병합
merged_data = pd.merge(dfa_30, dfb_30, on=['Month', 'Seq'])

# 결과 확인
print(merged_data)



# 결과 확인
# print(merged_df.head())
# print("병합된 데이터의 행 수:", merged_df.shape[0])



merged_data.to_csv("merged_2022_withbean.csv", index=False)
```
